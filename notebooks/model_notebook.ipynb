# Feature Engineering

```python
# Now, let's process the time based features
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['hour'] = df['trans_date_trans_time'].dt.hour
df['day'] = df['trans_date_trans_time'].dt.day
df['month'] = df['trans_date_trans_time'].dt.month
df['day_of_week'] = df['trans_date_trans_time'].dt.dayofweek
df['is_night'] = df['hour'].apply(lambda x: 1 if x >= 22 or x <= 6 else 0)
```

```python
# Extracting Age
df['dob'] = pd.to_datetime(df['dob']) # We are onverting dob to date time because the feature 'trans_date_trans_time' along with date also contains time. dob → contains only date, so after conversion it becomes a datetime with time = 00:00:00 (e.g., 1989-02-11 00:00:00)
df['age'] = (df['trans_date_trans_time'] - df['dob']).dt.days // 365
# You can see that we can extract the current age by subtracting current transaction date and dob.
```

```python
# Let's calculate distance between the merchant and customer.
from math import radians, sin, cos, sqrt, atan2
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # radius of Earth in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    return 2 * R * atan2(sqrt(a), sqrt(1-a))

df['distance_from_home'] = df.apply(lambda row:
    haversine(row['lat'], row['long'], row['merch_lat'], row['merch_long']), axis=1)
```

```python
# Transaction frequency & mean per card
df['avg_amt_per_card'] = df.groupby('cc_num')['amt'].transform('mean')

''' This gives the average transaction amount for each card.
It represents a card’s normal spending behavior.
💡 Why it helps:
If a card suddenly makes a transaction way above its usual average, that can indicate fraud.
Example: Card usually spends ₹500 → suddenly spends ₹50,000 → suspicious.'''

df['amt_to_mean_ratio'] = df['amt'] / df['avg_amt_per_card']

'''This measures how large the current transaction is relative to the user’s usual spending.
It’s a normalized feature — makes patterns comparable across all cards.
💡 Why it helps:
The ratio highlights deviation from typical spending.
Values around 1 → normal
Values much greater than 1 → abnormal → possibly fraud
Works better than raw amt, since everyone’s spending scale differs.
'''

df['transactions_per_day'] = df.groupby(['cc_num', 'day'])['amt'].transform('count')
```



## 🧠 Feature Encoding Strategy

In this dataset, several features are categorical in nature. Since machine learning models require numerical input, appropriate encoding techniques were applied based on the **cardinality** (number of unique categories) and the **relationship strength** of each feature with the target (`is_fraud`).
This ensures an optimal balance between **information preservation**, **dimensionality**, and **model performance**.

---

### 1️⃣ `gender` → **One-Hot Encoding**

**Reason:**

* The column `gender` has only two unique categories (`'M'` and `'F'`).
* One-hot encoding is ideal for such low-cardinality features, as it avoids implying any ordinal relationship between categories.

**Encoding Choice:**
👉 Convert into binary dummy variables (`gender_M`, `gender_F`).

**Why it works well for XGBoost:**
Tree-based models can easily handle binary splits (0/1) on such features.

---

### 2️⃣ `job` → **Target Encoding**

**Reason:**

* The `job` column has **high cardinality** (many unique occupations).
* Using one-hot encoding would create too many features and lead to overfitting or sparse data.
* Instead, each job category is replaced by the **mean fraud rate** for that job (computed on training data only).

**Encoding Choice:**
👉 Replace each category with its target mean:
[
\text{encoded_job} = E[\text{is_fraud} \mid \text{job}]
]

**Why it works well for XGBoost:**
Target encoding captures useful statistical relationships between categorical levels and the target variable without exploding feature space.

---

### 3️⃣ `merchant` → **Label Encoding / Frequency Encoding**

**Reason:**

* The `merchant` column has a **moderate number** of unique categories (hundreds to thousands).
* One-hot encoding would create too many columns, while target encoding might overfit (since merchant names can appear very few times).
* Label encoding (assigning each unique merchant an integer) or frequency encoding (replacing by how often each merchant appears) are efficient choices.

**Encoding Choice:**

* **Label Encoding** when category IDs are used directly.
* **Frequency Encoding** when representing how frequent a merchant is in the dataset.

**Why it works well for XGBoost:**
XGBoost can handle numerical encodings of categorical values naturally, as it finds optimal split thresholds, not linear relationships.

---

### 4️⃣ `category` → **One-Hot Encoding**

**Reason:**

* The `category` column has **moderate cardinality** (≈ 14 unique categories).
* Each category represents a distinct transaction type (e.g., `food_dining`, `travel`, `health_fitness`).
* One-hot encoding is appropriate because categories are non-ordinal and the number of unique values is manageable.

**Encoding Choice:**
👉 Create 14 binary dummy variables.

**Why it works well for XGBoost:**
Helps the model learn different fraud likelihoods per transaction type without implying order or magnitude between categories.

---

### ✅ Summary Table

| Column     | Unique Values | Encoding Type     | Reason                                                  |
| ---------- | ------------- | ----------------- | ------------------------------------------------------- |
| `gender`   | 2             | One-Hot Encoding  | Low cardinality, non-ordinal                            |
| `job`      | High          | Target Encoding   | High cardinality, directly related to fraud probability |
| `merchant` | Medium–High   | Label / Frequency | Avoid overfitting & maintain efficiency                 |
| `category` | 14            | One-Hot Encoding  | Moderate cardinality, clear categorical distinction     |

```python

from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from category_encoders import TargetEncoder
df_encoded = df.copy()
df_encoded = pd.get_dummies(df_encoded, columns=['gender'], drop_first=True)

target_encoder = TargetEncoder(cols=['job'])
df_encoded['job'] = target_encoder.fit_transform(df_encoded['job'], df_encoded['is_fraud'])

label_encoder = LabelEncoder()
df_encoded['merchant_label'] = label_encoder.fit_transform(df_encoded['merchant'])

merchant_freq = df_encoded['merchant'].value_counts() / len(df_encoded)
df_encoded['merchant_freq'] = df_encoded['merchant'].map(merchant_freq)
df_encoded.drop(columns=['merchant'], inplace=True)
df_encoded = pd.get_dummies(df_encoded, columns=['category'], drop_first=True)

```

```python
# Let's remove the unnecessary features
features_to_keep = [
    'amt', 'city_pop', 'job', 'hour', 'day', 'month', 'day_of_week',
    'is_night', 'age', 'distance_from_home', 'avg_amt_per_card',
    'amt_to_mean_ratio', 'transactions_per_day', 'gender_M',
    'merchant_label', 'merchant_freq'
] + [col for col in df.columns if col.startswith('category_')]+['is_fraud']
df=df_encoded[features_to_keep]
```

Splitting into train and test

```python
from sklearn.model_selection import train_test_split

X = df.drop('is_fraud', axis=1)
y = df['is_fraud']

x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

## Handling class Imbalance

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
x_train_res, y_train_res = smote.fit_resample(x_train, y_train)
```

## Scaling the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_res)
x_test_scaled = scaler.transform(x_test)
```

## Training with Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

lr = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
lr.fit(x_train_scaled, y_train_res)

y_pred_lor = lr.predict(x_test_scaled)
y_prob_lor = lr.predict_proba(x_test_scaled)[:, 1]

print(classification_report(y_test, y_pred_lor))
print("ROC-AUC Score:", roc_auc_score(y_test, y_prob_lor))
```

## Training with XGBoost

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=(y_train.value_counts()[0] / y_train.value_counts()[1]),
    random_state=42,
    n_jobs=-1
)

xgb.fit(x_train, y_train)

y_pred_xgb = xgb.predict(x_test)
y_prob_xgb = xgb.predict_proba(x_test)[:, 1]

print(classification_report(y_test, y_pred_xgb))
print("ROC-AUC Score:", roc_auc_score(y_test, y_prob_xgb))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Compute ROC points
fpr_lor, tpr_lor, _ = roc_curve(y_test, y_prob_lor)
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_prob_xgb)

# Compute AUCs
auc_lor = roc_auc_score(y_test, y_prob_lor)
auc_xgb = roc_auc_score(y_test, y_prob_xgb)

# Plot
plt.figure(figsize=(8,6))
plt.plot(fpr_lor, tpr_lor, label=f'Logistic Regression (AUC = {auc_lor:.4f})', lw=2)
plt.plot(fpr_xgb, tpr_xgb, label=f'XGBoost (AUC = {auc_xgb:.4f})', lw=2, color='orange')
plt.plot([0,1], [0,1], linestyle='--', color='gray')
plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import numpy as np

cm_lor = confusion_matrix(y_test, y_pred_lor)
cm_xgb = confusion_matrix(y_test, y_pred_xgb)

fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.heatmap(cm_lor, annot=True, fmt='d', cmap='Blues', cbar=False, ax=axes[0])
axes[0].set_title("Logistic Regression (LOR) Confusion Matrix")
axes[0].set_xlabel("Predicted")
axes[0].set_ylabel("Actual")

sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Oranges', cbar=False, ax=axes[1])
axes[1].set_title("XGBoost Confusion Matrix")
axes[1].set_xlabel("Predicted")
axes[1].set_ylabel("Actual")

plt.tight_layout()
plt.show()
```
